```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
import nltk
from nltk.corpus import wordnet as wn
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
text = 'Will you help me with the laundry?'
doc = nlp(text)
for token in doc:
    print(token.text, '->', token.pos_)
    if token.pos_ == 'VERB':
        new_text = WordNetLemmatizer().lemmatize(token.text, 'v')
        #print(new_text)
        m = text.replace(token.text , new_text)
        print(m)

    
```
